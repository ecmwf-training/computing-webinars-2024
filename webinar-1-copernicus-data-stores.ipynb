### Accessing data through the Data Store websites

In this example, we will download data from [ERA5](https://cds.climate.copernicus.eu/datasets/reanalysis-era5-single-levels-monthly-means?tab=overview) (the fifth generation ECMWF reanalysis for the global climate and weather). The [ERA5 monthly data on single levels from 1940 to present](https://cds.climate.copernicus.eu/datasets/reanalysis-era5-single-levels-monthly-means?tab=overview) dataset should be the first result.

Once you have reached the *ERA5 monthly data on single levels from 1940 to present* dataset, click on **Download** underneath the dataset title. This will take you to the download form, where you can manually select the data you are interested in.

As a simple example, let's explore some data which helps us understand how sea surface temperatures changed in the tropical Pacific Ocean during some El Niño and La Niña years, and compare these to a "normal" year.

With this in mind, let's download sea surface temperature data for the years 1993 (a relatively "normal" year), 1997 (an El Niño year) and 1998 (a La Niña year). Given that the effects of El Niño and La Niña on sea surface temperature are most clearly seen around December, let's download the data for December of each of these years.

Your selection should look something like this:

### Accessing data through the Python API

### Make a request with the CDS API
Now that we have set up our `.cdsapirc`, we can make a request for CDS data in a Python session. Let's make the same request we made earlier from the [ERA5 monthly data on single levels from 1940 to present](https://cds.climate.copernicus.eu/datasets/reanalysis-era5-single-levels-monthly-means?tab=overview) dataset - that is, for sea surface temperature for December of the years 1993, 1997 and 1998.

First of all, if you haven't already installed the cdsapi, uncomment and run the following cell:

```python
#!pip install cdsapi
```

Navigate to the [dataset page](https://cds.climate.copernicus.eu/datasets/reanalysis-era5-single-levels-monthly-means?tab=overview) and fill in the form again - or alternatively, find the request you made earlier in the [Your requests](https://cds.climate.copernicus.eu/requests?tab=all) page, then click *Details* followed by *Open request form*, which can be found next to the request ID.

Once you have completed the form, click the *Show API request code* box at the bottom of the page.


This should give you the following Python code, which you can run to programmatically download the data from the CDS:

```python
import cdsapi

dataset = "reanalysis-era5-single-levels-monthly-means"
request = {
    "product_type": "monthly_averaged_reanalysis",
    "variable": "2m_temperature",
    "year": ["1993", "1997", "1998"],
    "month": "12",
    "time": "00:00",
    "data_format": "grib",
    "download_format": "unarchived",
    "area": [20, 100, -20, 300]
}

client = cdsapi.Client()
client.retrieve(dataset, request).download()
```

### Accessing Data Store Services data with earthkit

In the previous sections, we explored how to access data from the Copernicus Data Store Services through the **web** and through the **Python API**. In both of these cases, we downloaded the data directly onto our machine as GRIB files.

In the following section, we will use **[earthkit](https://github.com/ecmwf/earthkit)** to construct a *workflow* of data access, processing and visualisation.

**earthkit** is a Python package which provides tools for accessing, processing and visualisation earth science data. It also offers direct access to the Copernicus Data Store Services, speeding up the process of getting data from the Data Stores into your Python session.

If you do not already have **earthkit-data** and **earthkit-plots** installed, uncomment and run the following cell. We will be using **earthkit-data** to access data from the CDS, and **earthkit-plots** to visualise it.

```python
#!pip install earthkit-data
#!pip install earthkit-plots
```

Now we can use the `from_source()` method from **earthkit-data** to make the same request we made earlier to the CDS API.

**earthkit-data** uses the same interface as the CDS API, with one difference - the first argument to `earthkit.data.from_source()` needs to be `"cds"` if we're accessing CDS data. After that, just like the `retrieve` method from the CDS API, we need to pass the name of the dataset followed by the request payload.

```python
import earthkit.data

dataset = "reanalysis-era5-single-levels-monthly-means"
request = {
    "product_type": "monthly_averaged_reanalysis",
    "variable": "sea_surface_temperature",
    "year": ["1993", "1997", "1998"],
    "month": "12",
    "time": "00:00",
    "area": [20, 100, -20, 300],
}

data = earthkit.data.from_source("cds", dataset, request)
```

One advantage of using **earthkit** instead of the CDS API is that the object returned from our request can be immediately opened and explored with commonly used scientific tools like `xarray`:

```python
data.to_xarray()
```

### Visualising the data

Another advantage of using **earthkit-data** to access Copernicus Data Store datasets is that the resulting object is understood by the full suite of earthkit tools, including **[earthkit-transforms](https://earthkit-transforms.readthedocs.io/en/latest/)** for data processing and **[earthkit-plots](https://earthkit-plots.readthedocs.io/en/latest/)** for visualisation.

To keep this example simple, let's use **earthkit-plots** to visualise the data we just retrieved from the CDS, and compare how sea surface temperatures in the tropical Pacific Ocean differ between El Niño and La Niña years.

```python
import earthkit.plots

YEARS = {
    "1993": "Normal conditions",
    "1997": "El Niño",
    "1998": "La Niña",
}

# Styles in earthkit-plots can include units
STYLE = earthkit.plots.styles.Style(
    colors="Spectral_r",
    levels=range(15, 33),
    units="celsius",
    ticks=range(15, 33),
)

figure = earthkit.plots.Figure(rows=3, size=(8, 7.5))

for i, year in enumerate(YEARS):
    subplot = figure.add_map(domain=[100, 300, -20, 20])
    subplot.contourf(data[i], style=STYLE)
    subplot.title(f"{YEARS[year]} - ERA5 {{short_name!u}} monthly mean: {{time:%B %Y}}")

figure.land(color="#555", zorder=2)
figure.gridlines(xstep=20, ystep=10, sharex=True)

figure.title("{variable_name} in {time:%B} of {time:%Y}", fontsize=14)

figure.legend(label="{variable_name!l} ({units})")
figure.show()
```
